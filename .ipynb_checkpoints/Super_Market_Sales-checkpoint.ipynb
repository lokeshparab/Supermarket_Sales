# Build  model

## Model 1

```python
df.head()
```

```python
df1 = df.drop(['Invoice ID','City','Customer type','Gender','Product line','DateTime','Payment','Weekday'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression  
```

```python
#preparinf the data
X = df1[['cogs','Ratings']].values 
y = df1['gross income'].values 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

```python

regressor = LinearRegression()  
regressor.fit(np.array(X_train).reshape(-1,1), y_train) 

print("Training complete.")
```

```python
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X.reshape(-1,1), y)
plt.plot(X.reshape(-1,1), line);
plt.show()
```

```python
y_pred = regressor.predict(np.array(X_test).reshape(-1,1))
```

```python
regressor.score(X_test.reshape(-1,1), y_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f'Mean absolute error: {mae:.2f}')
print(f'Mean squared error: {mse:.2f}')
print(f'Root mean squared error: {rmse:.2f}')
```
