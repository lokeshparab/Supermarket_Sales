```python
def weekday(no):
  days = ['Monday','Tueday','Wednesday','Thurday','Friday','Saturday','Sunday']
  return days[no]
```

# Build  model

```python
df1.head()
```

```python
df2 = df1.drop(['gross margin percentage','Branch','City','Gender','Product line','Date','Time','Payment','Customer type','Invoice ID'],axis=1)
df2
```

## Model 1

```python
#preparinf the data
X = df2['cogs'].values 
y = df2['gross income'].values 
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(np.array(X_train).reshape(-1,1), y_train) 

print("Training complete.")
```

```python
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X.reshape(-1,1), y)
plt.plot(X.reshape(-1,1), line);
plt.show()
```

```python
y_pred = regressor.predict(np.array(X_test).reshape(-1,1))
```

```python
regressor.score(X_test.reshape(-1,1), y_test)
```

```python
def hrs_fun(txt):
  data = txt.split()
  time = data[1].split(':')
  hrs = int(time[0])
  if hrs > 12:
    return str( hrs - 12 )+ ':' + ':'.join(time[1:]) + 'pm'
  else:
    return data[1]+'am'

time_= '2022-06-18 17:42:52.902'    
print(hrs_fun(time_))
```
